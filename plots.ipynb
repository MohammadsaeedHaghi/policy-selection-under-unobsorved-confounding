```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
obj_primal = pd.read_csv('objective_values_primal.csv')
obj_dual = pd.read_csv('objective_values_dual.csv')
```

```python
obj_primal = pd.read_csv('objective_values_primal.csv')
obj_dual = pd.read_csv('objective_values_dual.csv')

plt.figure(figsize=(10, 6))

plt.plot(obj_primal['Gamma'], obj_primal['Objective_Value'], 
         marker='o', linestyle='-', label='Primal Objective', 
         color='blue', alpha=0.7)

plt.plot(obj_dual['Gamma'], obj_dual['Objective_Value'], 
         marker='s', linestyle='-', label='Dual Objective', 
         color='red', alpha=0.7)

plt.xlabel('Gamma')
plt.ylabel('Objective Value')
plt.title('Primal and Dual Objective Values vs Gamma')
plt.legend()
plt.grid(True)

plt.show()
```
